```python
import shapefile as shp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sf = shp.Reader("GeoData/Regioni/Reg2015_WGS84.shp")

plt.figure()
for shape in sf.shapeRecords():
    x = [i[0] for i in shape.shape.points[:]]
    y = [i[1] for i in shape.shape.points[:]]
    plt.plot(x,y)
plt.show()
```
